```python
#AGGREGATED - TRANSACTION DATA
#This is to direct the path to get the data as states
path="C:/Users/mosel/Documents/PYTHON/project/project_phonepe_transaction_insight/dataset/data/aggregated/transaction/country/india/state/"
#Agg_state_list--> to get the list of states in India
Agg_state_list=os.listdir(path)
Agg_state_list

#This is to extract the data's to create a dataframe
clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Created a dataframe
aggregated_transaction=pd.DataFrame(clm)
aggregated_transaction
```

```python
#AGGREGATED - USERS 
#This is to direct the path to get the data as states
path="C:/Users/mosel/Documents/PYTHON/project/project_phonepe_transaction_insight/dataset/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

#This is to extract the data's to create a dataframe
clm={'State':[], 'Year':[],'Quater':[],'Users_brand':[], 'Users_count':[], 'Users_percentage':[], 'RegisteredUsers':[], 'AppOpens':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            aggregated_data = D.get('data', {}).get('aggregated', {})
            registeredUsers = aggregated_data.get('registeredUsers', 0)
            appOpens = aggregated_data.get('appOpens', 0)

            if D['data'].get('usersByDevice') is not None:
              for z in D['data']['usersByDevice']:
                brand = z.get('brand', 'Unknown')
                count = z.get('count', 0)
                percentage = z.get('percentage', 0.0)
                clm['Users_brand'].append(brand)
                clm['Users_count'].append(count)
                clm['Users_percentage'].append(percentage)
                clm['RegisteredUsers'].append(registeredUsers)
                clm['AppOpens'].append(appOpens)

                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quater'].append(int(k.strip('.json')))
#Created a dataframe
aggregated_users=pd.DataFrame(clm)
aggregated_users
```
